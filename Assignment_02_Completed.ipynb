```python
#Importing the commonly used libraries

import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
#Question 01
#Comparing the information of top 10 most populated countries using line chart
df = pd.read_csv('/world_pop.csv')

# Extract population data for the top 10 countries
top_10_countries_2020 = df[['country', 'year_2020']].sort_values(by='year_2020', ascending=False).head(10)

top_10_populations = df[df['country'].isin(top_10_countries_2020['country'])]
top_10_populations.set_index('country', inplace=True)

# Plot the line chart
plt.figure(figsize=(12, 8))

for country in top_10_populations.index:
    years = [int(year.split('_')[1]) for year in top_10_populations.columns]
    plt.plot(years, top_10_populations.loc[country], label=country)

plt.title('Population of Top 10 Highest Populated Countries (1960-2020)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(loc='upper left', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Question 02

#Part a. Find the least popoulous coutnries in 2015
df_2020 = df[['country', 'year_2015']]
df_2020.dropna(subset=['year_2015'], inplace=True)

# Sort the DataFrame by population in descending order
df_2020_sorted = df_2020.sort_values(by='year_2015', ascending=True)
least_10_countries = df_2020_sorted.head(10)

# Extract country names and populations from the DataFrame
countries = least_10_countries['country']
populations = least_10_countries['year_2015']

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.barh(countries, populations, color='pink')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top 10 Least Populated Countries in 2015')
plt.gca().invert_yaxis()  # Invert y-axis to display countries with lowest population at the top
plt.show()
```

```python
#Question 02

#Part b.Find the popoulation change of PK,US,UK and India
# List of specific countries you want to extract population information for
countries_1 = ['Pakistan', 'United States', 'United Kingdom', 'India']

# Filter the DataFrame to include only the specified countries
population_data = df[df['country'].isin(countries_1)]
population_data.dropna(subset=[f'year_{year}' for year in range(1960, 2021)], inplace=True)

# Plot the line chart for each country
plt.figure(figsize=(10, 6))  # Set the figure size

for country in countries_1:
    plt.plot(population_data.columns[1:], population_data.loc[population_data['country'] == country].values[0][1:], label=country)

plt.title('Population of Specific Countries (1960-2020)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Question 02

#Part c. Find PK population growth from 2010-2020
country_PK = ['Pakistan']

population_data = df[df['country'].isin(country_PK)]
population_data.dropna(subset=[f'year_{year}' for year in range(1960, 2020)], inplace=True)

# Extract the population data for the specified country
years = range(2010, 2020)
max_population = [population_data[f'year_{year}'].max() for year in years]

# Plot the line chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(years, max_population, color = 'pink')
plt.title('Maximum Population of Pakistan (1960-2020)')
plt.xlabel('Year')
plt.ylabel('Maximum Population')
plt.ticklabel_format(style='plain', axis='y')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Question 03
#Show the relationship between carats and prices for all diamonds with SI2 clarify and E Color

dataset = pd.read_csv('/diamonds.csv')
filtered_df = dataset[(dataset['clarity'] == 'SI2') & (dataset['color'] == 'E')]

# Plot the filtered DataFrame
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['carat'], filtered_df['price'], alpha=0.5, color = 'red', label = 'carat vs price')
plt.title('Scatter Plot of Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Question 04
#Use Folium to represent nuclear waste sites in US

US_center = 37.0902, -95.7129
US_map = folium.Map(location=US_center, zoom_start = 4)

#Add location markers to the map
nuclear_df = pd.read_csv('/nuclear_waste_sites.csv')
for i, row in nuclear_df.iterrows():
    folium.Marker(location=[row['lat'], row['lon']],popup=row['text']).add_to(US_map)

#show the map
US_map
```
